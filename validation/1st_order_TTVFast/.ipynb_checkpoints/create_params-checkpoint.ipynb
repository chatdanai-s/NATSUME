This file aims to create TTV signals to validate 1st order MMR code with regimes
- Eccentricities: 0.01 to 0.3 / Uniform log e
- Longitudes of periastron: 0 to 360 deg / Uniform w
- MMRs: 2:1, 3:2, 4:3, 5:4 / Uniform
- Deltas: Depends on j / Uniform
- Planet masses: 0.3 to 300 Earth masses / Uniform log m

If system with invalid regime e >~ Delta^2 / mu is detected, then DO NOT SIMULATE.

```python
# Evaluate Delta boundary for each j
def periodRatio(j):
    return j/(j-1)

def minmaxDelta(j):
    if j == 2:
        maxPeriodRatio = np.nan
        maxDelta = 0.2
    else:
        maxPeriodRatio = 0.5 * (periodRatio(j) + periodRatio(j-1))
        maxDelta = (j-1)/j * maxPeriodRatio - 1

    minPeriodRatio = 0.5 * (periodRatio(j) + periodRatio(j+1))
    minDelta = (j-1)/j * minPeriodRatio - 1
    
    return minDelta, maxDelta

minmaxDeltas = pd.DataFrame(
    [[j, *minmaxDelta(j)] for j in (2, 3, 4, 5)],
    columns=['j', 'min_Delta', 'max_Delta']
)

# Build lookup dictionaries once for Dynamic e/Delta sampling
minmaxDeltas['min_Delta_used'] = [0.020, 0.010, 0.005, 0.005]
minmaxDeltas['max_Delta_used'] = [0.050, 0.020, 0.015, 0.010]
minDelta_lookup = dict(zip(minmaxDeltas['j'], minmaxDeltas['min_Delta_used']))
maxDelta_lookup = dict(zip(minmaxDeltas['j'], minmaxDeltas['max_Delta_used']))

minmaxDeltas['min_Loge_used'] = [-2, -2, -2, -2]
minmaxDeltas['max_Loge_used'] = [-0.7, -0.7, -1, -1]
minLoge_lookup = dict(zip(minmaxDeltas['j'], minmaxDeltas['min_Loge_used']))
maxLoge_lookup = dict(zip(minmaxDeltas['j'], minmaxDeltas['max_Loge_used']))

# Show df
minmaxDeltas.head()
```

```python
# Create 3000 samples per MMR
N = 3000
mstar = 1
P1 = 7
validity_threshold = 100

jlist = [2, 3, 4, 5]
rows = []

for i,j in enumerate(jlist):
    while len(rows) < (i+1)*N:
        Loge_low = minLoge_lookup[j]
        Loge_high = maxLoge_lookup[j]
        e1 = 10 ** np.random.uniform(Loge_low, Loge_high)
        e2 = 10 ** np.random.uniform(Loge_low, Loge_high)
        w1 = np.random.uniform(0, 360)
        w2 = np.random.uniform(0, 360)

        Delta_low = minDelta_lookup[j]
        Delta_high = maxDelta_lookup[j]
        Delta = np.random.uniform(Delta_low, Delta_high) * np.random.choice([-1, 1])
        P2 = P1 * (Delta + 1) * j / (j-1)
    
        m1 = 10 ** np.random.uniform(-0.5, 2.5)
        m2 = 10 ** np.random.uniform(-0.5, 2.5)
        mu1 = (m1 * u.M_earth).to(mstar * u.M_sun).value
        mu2 = (m2 * u.M_earth).to(mstar * u.M_sun).value
    
        e_max = max(e1, e2)
        validity1 = Delta**2 / (mu2 * e_max)
        validity2 = Delta**2 / (mu1 * e_max)
    
        if (validity1 > validity_threshold) and (validity2 > validity_threshold):
            rows.append([
                e1, w1, e2, w2, Delta, j, P1, P2,
                m1, m2, mu1, mu2, validity1, validity2
            ])

df = pd.DataFrame(
    rows,
    columns=[
        'e1','w1','e2','w2',
        'Delta','j','P1','P2',
        'm1','m2','mu1','mu2',
        'validity1','validity2'
    ]
)

df['valid'] = (validity1 > validity_threshold) & (validity2 > validity_threshold)
valid_percent = df['valid'].mean() * 100
df.head()
```

```python
# Histogram of validity condition
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
fmt = lambda x: f"{x:.2f}".rstrip("0").rstrip(".")
labels = [f"Valid systems ({fmt(valid_percent)}%)", f"Invalid systems ({fmt(100-valid_percent)}%)"]
colors = ["royalblue", "tomato"]
xlim = (2, 6); ylim = (0, 700)

v1 = df['validity1'].to_numpy()
v2 = df['validity2'].to_numpy()
valid = (df['valid'] == True).to_numpy()

# Left histogram
ax[0].hist([np.log10(v1[valid]), np.log10(v1[~valid])],
           bins=50, stacked=True, label=labels, color=colors)
ax[0].set_xlabel('log$_{10}$($\Delta^2 / \mu_1 e$)')
ax[0].set_ylabel('Count')
ax[0].set_xlim(xlim); ax[0].set_ylim(ylim)
# ax[0].legend(loc='upper left')

# Right histogram
ax[1].hist([np.log10(v2[valid]), np.log10(v2[~valid])],
           bins=50, stacked=True, label=labels, color=colors)
ax[1].set_xlabel('log$_{10}$($\Delta^2 / \mu_2 e$)')
ax[1].set_ylabel('Count')
ax[1].set_xlim(xlim); ax[1].set_ylim(ylim)
# ax[1].legend(loc='upper left')

plt.tight_layout()
plt.show()
```
